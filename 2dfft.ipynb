```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path="dataset/datachunks/channel1_3m_train0_5.wav"
file,sample_rate=librosa.load(file_path,sr=None)
```

```python
window_size=int(512)
wd = np.hamming(window_size)
slide_size = int(4)
overlap = window_size - slide_size
spectrum,frequency,time,im=plt.specgram(file,NFFT=window_size,Fs=sample_rate,window=wd,noverlap=overlap,mode='magnitude')
plt.colorbar(label="magnitude")
plt.xlabel("time")
plt.ylabel("frequency")
plt.show()
```
